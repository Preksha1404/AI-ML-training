### Analyze which game has the highest global sales

```python
highest_global_sales=df[df.Global==df.Global.max()]
highest_global_sales.Game
```

### Analyze which games have the lowest sales in Japan.

```python
lowest_sales_in_japan=df[df.Japan==df.Japan.min()]
lowest_sales_in_japan.Game
```

### Analyze how much sales "Grand Theft Auto V" made globally

```python
df[df.Game=="Grand Theft Auto V"].Global
```

### Analyze which games released in 2015 have the highest sales globally. 

```python
maximum_sales_by_year=df.groupby('Year')[['Game','Global']].max().reset_index()
maximum_sales_by_year[maximum_sales_by_year['Year']==2015]
```

### Analyze which "Action" games were released in 2016. 

```python
data=df[df['Genre']=='Action']
data[data.Year==2016]
```

### Analyze which game from "Activision" made the best sales in "North America".

```python
### Analyze which game from "Activision" made the best sales in "North America".
data=df.groupby('Publisher')[['Game','Genre','Publisher','North America']].max()
data[data['Publisher']=='Activision']
```

### Analyze what the total global sales are for each year.

```python
### Analyze what the total global sales are for each year.
yearly_sales=df.groupby('Year')['Global'].sum()
yearly_sales
```

### Analyze which are the top 3 companies in total global sales.

```python
 ### Analyze which are the top 3 companies in total global sales.
sales_by_publisher = df.groupby('Publisher')['Global'].max()
top_publishers = sales_by_publisher.sort_values(ascending=False).head(3)
top_publishers
```
