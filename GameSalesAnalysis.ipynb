```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Plot a bar graph showing sales of the "Spider-Man (PS4)" game in different regions with proper titles and lines. 

```python
game_data = df[df['Game'] == 'Spider-Man (PS4)'].iloc[0]

regions = ['North America', 'Europe', 'Japan', 'Rest of World', 'Global']

region_sales = game_data[regions]

# Create the bar chart
plt.bar(regions, region_sales)
plt.show()
```

## Plot a line plot to compare the global sales made through different years by the following companies (Activision, EA Sports, Sony Interactive Entertainment, Namco Bandai Games, Capcom, Warner Bros. Interactive)

```python
df
```

```python
# Define companies to analyze
companies = ['Activision', 'EA Sports', 'Sony Interactive Entertainment','Namco Bandai Games', 'Capcom', 'Warner Bros. Interactive']

# Filter dataframe for selected companies
companies_df = df[df['Publisher'].isin(companies)]
# companies_df

# Group by Year and Publisher, then sum the Global sales
sales_by_year = companies_df.groupby(['Year', 'Publisher'])['Global'].sum().reset_index()
# sales_by_year

# Create a pivot table which display total global sales by year and publisher
pivot_data = sales_by_year.pivot(index='Year', columns='Publisher', values='Global')
# pivot_data

# Plot the data
plt.figure(figsize=(12, 6))
pivot_data.plot(marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.title('Global Sales by Publisher Over Time')
plt.legend(title='Publisher')

# Show the plot
plt.show()
```

## Plot a pie chart for the global sales made in 2018 by the companies (Activision, EA Sports, Sony Interactive Entertainment, Namco Bandai Games, Capcom, Warner Bros. Interactive).

```python
# Define companies to analyze
companies = ['Activision', 'EA Sports', 'Sony Interactive Entertainment','Namco Bandai Games', 'Capcom', 'Warner Bros. Interactive']

# Filter dataframe for selected companies
companies_df = df[df['Publisher'].isin(companies)]
# companies_df

sales_by_year = companies_df.groupby(['Year', 'Publisher'])['Global'].sum().reset_index()
sales_by_specific_year=sales_by_year[sales_by_year['Year']==2018]

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(
        sales_by_specific_year['Global'],
        labels=sales_by_specific_year['Publisher'],
        autopct='%1.1f%%',
        startangle=90,
    )
plt.title('Global Sales Distribution Among Gaming Companies in 2018')
plt.show()
```
