```python
df.fillna({"Publisher":"unknown"},inplace=True)
```

```python
df.fillna(df.Year.mode()[0],inplace=True)
```

```python
df.drop(['Rank','Name','Platform'], axis=1,inplace=True)
```

```python
df.head()
```

```python
# One HotEncoding-->using get_dummies
df=pd.get_dummies(data=df,columns=['Publisher','Genre'],drop_first=True)
```

```python
df.head()
```

```python
X = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
# K value
# Elbow method --> Manual way --> Initialize K values and find wcss and select K value
wcss=[]

for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot Elbow method plot
plt.plot(range(1,11),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
from kneed import KneeLocator
# If wcss values are increasing --> curve=cocave and direction=increasing
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow # Find the exact value of K
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
df['Cluster']=kmeans.fit_predict(X_scaled)
```

```python
df
```

```python
plt.scatter(df['NA_Sales'], df['EU_Sales'], c=df['Cluster'], cmap='viridis')
plt.xlabel("NA Sales")
plt.ylabel("EU Sales")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X_scaled,df['Cluster'])
print(score)
```
